```python
import numpy as np
```

```python
h = np.genfromtxt(r"C:\Users\heena\Downloads\Iris_clean_-_Copy.csv" ,dtype="float16",delimiter = ",")
h
```

```python
sepal_length = h[1:,0:1]
sepal_length
```

```python
np.where(sepal_length>10) # outliers
```

```python
sepal_length = np.where(sepal_length>10,np.nanmedian(sepal_length),sepal_length) 
sepal_length
```

```python
sepal_length = np.where(np.isnan(sepal_length),np.nanmean(sepal_length),sepal_length)
sepal_length
```

```python
sepal_width = h[1:,1:2]
sepal_width
```

```python
np.where(sepal_width>5)
```

```python
sepal_width = np.where(np.isnan(sepal_width),np.nanmean(sepal_width),sepal_width)
sepal_width                      
```
